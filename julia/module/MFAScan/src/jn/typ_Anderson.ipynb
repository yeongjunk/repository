```julia
include("../TAU_3.jl")
using Random
using LatticeModels
using Plots
using ProfileView
plotly()
```

```julia
struct CubicParams
    L::Int
    W::Float64
end
```

```julia
function f(p::CubicParams; rng = Random.GLOBAL_RNG)
    H = cubic(L = p.L, M = p.L, N = p.L) .+ p.W*Diagonal(rand(L*L*L) .- 0.5)
end

function g(p::CubicParams; rng = Random.GLOBAL_RNG)
    H = square(M = p.L, N = p.L) .+ p.W*Diagonal(rand(L*L) .- 0.5)
end

function h(p::CubicParams; rng = Random.GLOBAL_RNG)
    H = chain(N = p.L) .+ p.W*Diagonal(cos.(2π*(1:L)*(1 + √5)/2 .+ 2*pi*rand(rng)))
end

```

```julia
L = 50; W = 2.0; E_c = 0.; E_del = 0.3
l = [1, 2, 5, 10]
q = collect(0.:0.1:3.)
p = CubicParams(L, W)
ltc = Lattice2D(L, L, 1)
```

```julia
E, gipr, τ, α, f_α = scan_ταf(g, p, E_c, E_del, ltc, L, l=l, q=q, R=1, nev = 100, seed = rand(1:19999))
```

```julia
plot!((l./L), τ', marker = :circle)
```

```julia
plot!((l./L)[2:end], diff(log10.(gipr'), dims=1)./diff(log10.(l./L)), marker = :circle)
```

```julia
p1 = plot(q, τ, legend = false)
p2 = plot(q, α, legend = false)
p3 = plot(α, f_α)
p4 = plot(p1, p2, p3)
```
