```julia
using Revise
using LatticeModels
using LinearAlgebra
using Plots
```

```julia
M = 20; N = 20
H = kagome(N = N, M = M);
```

```julia
vals, vecs = eigen(Matrix(H));
```

```julia
plot(vals)
```

```julia
psi_p = vecs[:, M*N*2:end];
```

```julia
P = zeros(size(psi_p, 1), size(psi_p, 1))
for i in 1:size(psi_p, 2)
    P += psi_p[:, i]*psi_p[:, i]'
end
```

```julia
scatter(log.(abs.(P[1, 1:3:3N÷2])), xaxis = :log)
```
