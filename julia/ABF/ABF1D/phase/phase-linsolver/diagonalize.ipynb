```julia
using Plots
using ABF
using Lattice
include("./phase.jl")
```

```julia
function phase_dis(H; V = 1., rng = nothing)
    F = eltype(H)
    D = convert.(Complex{F}, H)
    if rng == nothing
        rng = MersenneTwister()
    end
    rows = rowvals(D)
    vals = nonzeros(D)
    m, n = size(D)
    for j = 1:n
       for i in nzrange(D, j)
          row = rows[i]
          # println("$row ,", "$j")
          if row > j
              vals[i] = im*V*vals[i]*(rand(rng, F) .- 0.5)
          elseif row <= j
              vals[i] = F(0.)
          end
       end
    end
    return D + D'
end
```

```julia
L = 201
r = rand(1:10000)
H = ham_sf_pbc(L = L, rng = MersenneTwister(r))
H2 = ham_sf_obc(L = L, rng = MersenneTwister(r))
```

```julia
H2, U = ham_fe_obc(Lattice1D(L, 2), big(-2.), big(0.), big(0.25))
D = phase_dis(H2)

H_sf = ComplexF64.(project(U'*D*U))
```

```julia
vals, vecs = eigen(Hermitian(Matrix(H)), L÷2:L÷2+2);
vals2, vecs2 = eigen(Hermitian(Matrix(H_sf)), L÷2:L÷2+2);
```
