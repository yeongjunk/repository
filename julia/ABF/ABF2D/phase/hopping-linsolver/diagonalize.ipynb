```julia
using Plots
using Random
using Lattice
include("./hopping.jl")
default(format=:png)
using ColorSchemes
using LaTeXStrings
using CSV, DataFrames
```

```julia
function latticeplot2D(ltc, psi)
    ms = Int64[]
    ns = Int64[]
    for i in 1:length(psi)
        m,n, _ = site.(Ref(ltc), i)
        push!(ms, m)
        push!(ns, n)
    end
    mn = hcat(ms, ns)
    p = scatter(ms, ns, marker_z = abs.(psi), 
        msw = 0, c = :lajolla, frame = :box, label = false, 
        markersize = 6, colorbar_title = L"\psi",  thickness_scaling = 1)
    xlabel!("x")
    ylabel!("y")
    return p
end

function latticeplot2d_data(ltc, psi)
    ms = Int64[]
    ns = Int64[]
    for i in 1:length(psi)
        m,n, _ = site.(Ref(ltc), i)
        push!(ms, m)
        push!(ns, n)
    end
    mn = hcat(ms, ns)
    df = DataFrame(x = ms, y=ns, psi_abs = abs.(psi))
    return df
end
```

```julia
L = 50
ltc = Lattice2D(L, L, 1)
H = square_pbc(ltc)
H2 = square_obc(ltc)
```

```julia
vals, vecs = eigen(Hermitian(Matrix(Float64.(H))), L^2÷2:L^2÷2+5);
vals2, vecs2 = eigen(Hermitian(Matrix(Float64.(H2))), L^2÷2:L^2÷2+5);
ltc = Lattice2D(L, L, 1)
```

```julia
display(vals)
latticeplot2D(ltc, vecs[:, 3])
```

```julia
display(vals2)
idx = findmin(abs.(vals2))
p = latticeplot2D(ltc, vecs2[:, 1])
```

```julia
df = latticeplot2d_data(ltc, vecs2[:, 1])
```
