```julia
using Plots
using Random
include("./phase.jl")
include("./gr.jl")
```

```julia
L = 40
r = rand(1:10000)
H = ham_sf_pbc(L = L, θ = big(0.25), rng = MersenneTwister(r))
H2 = ham_sf_obc(L = L, θ = big(0.25), rng = MersenneTwister(r))
```

```julia
H2[101,:]
```

```julia
vals, vecs = eigen(Hermitian(Matrix(ComplexF64.(H))), L÷2:L÷2+2);
vals2, vecs2 = eigen(Hermitian(Matrix(ComplexF64.(H2))), L÷2:L÷2+2);
```

```julia
plot((abs.(vecs[:, 2])), label = "PBC")
plot!((abs.(vecs2[:, 2])), label = "OBC")
```

```julia
plot(log.(abs.(vecs[:, 2])), label = "PBC")
plot!(log.(abs.(vecs2[:, 2])), label = "OBC")
```

```julia
size(vecs)
```

```julia
lw = LatticeWave(Lattice2D(L, L, 1), log.(abs.(vecs2[:, 2])))
```

```julia
eig_corr(lw, (5, 4), direction = :x)
```

```julia
gr = filter(!isnan, eig_corr_full(lw))
```

```julia
plot(gr, yaxis = :log10, xaxis = :log10)
```
