```julia
using Plots
using LinearAlgebra
using DelimitedFiles
default(format =:png)
```

```julia
data = readdlm("./gr.dat");
```

```julia
p = plot()
plot!(p, data.^2)

display(p)
ylabel!("g(r)^2")
xlabel!("r")
```

```julia
xis = Float64[]
for i in 1:21
    lyap = (data[i, end] - data[i, end-10])/11
    push!(xis, 1/lyap)
end
scatter(xis)
```
