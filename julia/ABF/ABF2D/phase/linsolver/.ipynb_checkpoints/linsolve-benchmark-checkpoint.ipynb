```julia
using Plots
include("./phase.jl")
```

```julia
L = 2001
r = rand(1:10000)
H = ham_sf_pbc(L = L, rng = MersenneTwister(r))
H2 = ham_sf_obc(L = L, rng = MersenneTwister(r))
```

```julia
vals, vecs = eigen(Hermitian(Matrix(H)), L÷2:L÷2+2);
vals2, vecs2 = eigen(Hermitian(Matrix(H2)), L÷2:L÷2+2);
```

```julia
plot((abs.(vecs[:, 2])), label = "PBC")
plot!((abs.(vecs2[:, 2])), label = "OBC")
```

```julia
plot(log.(abs.(vecs[:, 2])), label = "PBC")
plot!(log.(abs.(vecs2[:, 2])), label = "OBC")
```
