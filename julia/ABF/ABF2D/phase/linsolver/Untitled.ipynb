```julia
using Plots
using LinearAlgebra
using DelimitedFiles
default(format =:png)
```

```julia
data = readdlm("./gr.dat");
```

```julia
p = plot()
r = 1:length(data)
datasq = data.^2
plot!(p, datasq, legend = false, lw = 2)
ylabel!("g(r)^2")
xlabel!("r")
slope = (datasq[end]-datasq[1])/length(data)
plot!(p, slope*(1:length(data)), ls = :dot, c = :black, lw = 2)

savefig(p, "p.pdf");
display(p)
```

```julia
p1 = plot()
r = 1:length(data)
datasq = data.^2
plot!(p1, data .+ 1, legend = false, lw = 2, c = :red, yaxis = :log10, xaxis = :log10)
ylabel!("g(r)")
xlabel!("r")
```
